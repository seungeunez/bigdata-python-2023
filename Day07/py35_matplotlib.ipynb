##### matplotlib 사용

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
arr = np.random.randn(30)
arr
````

````python
plt.plot(arr)
````

````python
# 다중차트
fig = plt.figure() # Figure 객체 생성
ax1 = fig.add_subplot(2, 2, 1) #차트1 (2행 2열로 차트 중 1)
ax2 = fig.add_subplot(2, 2, 2) #차트2 (2행 2열로 차트 중 2)
ax3 = fig.add_subplot(2, 2, 4) #차트3 (2행 2열로 차트 중 4)

ax1.hist(np.random.randn(100)) #분포표
ax2.scatter(np.arange(30), np.random.randn(30)) # 스캐터차트
ax3.plot(np.random.randn(50).cumsum()) # 라인차트

plt.show() # 차트보기
````

````python
# 안쓰는 차트도 같이 나옴
fig, axes = plt.subplots(2, 2)
axes[0][0].hist(np.random.randn(100)) #분포표
axes[0][1].scatter(np.arange(30), np.random.randn(30)) # 스캐터차트
axes[1][1].plot(np.random.randn(50).cumsum()) # 라인차트

plt.show()
````

##### 차트꾸미기

````python
# 차트색상, 라인종류, 마커
fig, axes = plt.subplots(2, 2)
axes[0][0].hist(np.random.randn(100), bins=20 ,color='red', alpha=1.0) #bins = 범위 20개로 늘림, alpha = 투명도(0,0[투명]~1.[불투명])
axes[0][1].scatter(np.arange(30), np.random.randn(30), color='lightgreen') 
axes[1][0].plot(np.random.randn(50).cumsum(), linestyle='--', color='red', marker='1') #marker = 값위치표시 *(별), ^(삼각형), o, 1
axes[1][1].plot(np.random.randn(50).cumsum(), linestyle=':', marker='o')#linestyle = 선종류 -(일반선) --(대시), :(점), -.(선점) 

plt.show()
````

````python
# 범례 legend
plt.plot(np.random.randn(50).cumsum(), label='x')
plt.plot(np.random.randn(50).cumsum(), label='y')
plt.legend(loc='best') # 'best', 'upper left', 'upper right', 'lower left', 'lower right', 'right', 'center'

plt.show()
````

##### Pandas 데이터 프레임, 시리즈

````python
# 판다스 시리즈
pd.Series(np.random.randn(1000).cumsum()).plot()
plt.show()
````

````python
# 데이터 프레임 plot, his
pd.DataFrame(np.random.randn(10, 4).cumsum(0), columns=['A', 'B', 'C', 'D'], index=np.arange(0, 100, 10)).plot()
````
