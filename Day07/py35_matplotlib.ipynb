##### matplotlib 사용

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
arr = np.random.randn(30)
arr
````

````python
# 단일차트
plt.plot(arr)
````

````python
# 다중차트
fig = plt.figure() # Figure 객체 생성
ax1 = fig.add_subplot(2, 2, 1) # 차트1(2행 2열 차트중 1)
ax2 = fig.add_subplot(2, 2, 2) # 차트2(2행 2열 차트중 2)
ax3 = fig.add_subplot(2, 2, 4) # 차트3(2행 2열 츠트중 3)

ax1.hist(np.random.randn(100)) # 분포표
ax2.scatter(np.arange(30), np.random.randn(30)) # 스캐터차트
ax3.plot(np.random.randn(50).cumsum()) # 라인차트

plt.show() # 차트 보기
````

````python
# 안쓰는 차트도 같이 나옴
fig, axes = plt.subplots(2, 2)
axes[0][0].hist(np.random.randn(100)) # 분포표
axes[0][1].scatter(np.arange(30), np.random.randn(30)) # 스캐터차트
axes[1][1].plot(np.random.randn(50).cumsum()) # 라인차트

plt.show()
````

##### 차트꾸미기

````python
# 차트색상, 라인종류, 마커
fig, axes = plt.subplots(2, 2)
axes[0][0].hist(np.random.randn(100), bins=20, color='red', alpha=0.5) # bins = 범위 20개로 늘림, alpha = 투명도(0.0[투명]~1.0[불투명])
axes[0][1].scatter(np.arange(30), np.random.randn(30), color='lightgreen')
axes[1][0].plot(np.random.randn(50).cumsum(), linestyle=':', color='red', marker='1') # marker = 값위치표시 *(별), ^(삼각형), o, 1
axes[1][1].plot(np.random.randn(50).cumsum(), linestyle='--', marker='o') # linestyle = 선종류 -(일반선) --(대시), :(점), -.(선점)

plt.show()
````

````python
# 범례 legend
plt.plot(np.random.randn(50).cumsum(), label='x')
plt.plot(np.random.randn(50).cumsum(), label='y')
plt.legend(loc='upper right') # 'best' 'upper left', 'upper right', 'lower left','lower right', 'right', 'center'

plt.show()
````

##### Pandas 데이터프레임, 시리즈

````python
# 판다스 시리즈
pd.Series(np.random.randn(1000).cumsum()).plot()
````

````python
# 한글 깨짐현상
import matplotlib
from matplotlib import font_manager, rc

font_path = 'C:/Windows/Fonts/NanumGothic.ttf' 
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font) # 한글깨짐현상 해결
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 사용시 마이너스 깨짐현상 해결

# x, y축 설정
chart = pd.Series(np.random.randn(1000).cumsum()).plot()
chart.set_xticks([0, 250, 500, 750, 1000]) # x축 간격 조정
# chart.set_yticks([0, 25, 50]) # y축 간격 조정
chart.set_title('랜덤 Linechart') # 영어는 잘되는데 한글은 기본 깨짐
chart.set_xticklabels(['zero', 'one', 'two', 'three', 'four'], rotation=300) # x축을 라벨로 표시
chart.grid(True) # 그리드선 표시

plt.show() # 쓸데없는 타입, 결과 메시지 삭제
````

#### 예제 연습

캐글 타이타닉 연습문제 CSV

https://raw.githubusercontent.com/developer-sdk/kaggle-python-beginner/master/datas/kaggle-titanic/train.csv

````python
train = pd.read_csv('https://raw.githubusercontent.com/developer-sdk/kaggle-python-beginner/master/datas/kaggle-titanic/train.csv')

train.tail()
````

````python
# 성별 생존률 교차도표 생성
gender_survived = pd.crosstab(train['Sex'], train['Survived'])
gender_survived.columns = gender_survived.columns.map({0:"Dead", 1:"Alive"})
````

````python
# 바차트
# bar = gender_survived.plot.bar()
bar = gender_survived.plot(kind='bar', title='생존자/사망자')

# 차트의 수치표시
for p in bar.patches:
    left, bottom, width, height = p.get_bbox().bounds
    bar.annotate(f'{height:n}', (left+width/2, height*1.02), ha='center')

gender_survived.plot.bar(subplots=True) # 성별로 따로 바차트를 추가 생성
# bar.set_title('생존자/사망자수')
# gender_survived.plot.bar(x='Alive') # 차트는 사망자, x축에는 생존자수치 표시(차트와 x라벨을 교차사용 : 쓸일 없음)
# gender_survived.plot.bar(x='Dead')
# gender_survived.plot.bar(y='Dead')

plt.show()
````

````python
# 가로 바차트
# gender_survived.plot(kind='barh')
gender_survived.plot.barh()
gender_survived.plot.barh(subplots=True)
plt.show()
````

````python
# 라인차트
gender_survived.plot(kind='line')
````

````python
# 파이차트
gender_survived['Alive'].plot(kind='pie', title='생존자 비율', autopct='%1.1f%%')
````
