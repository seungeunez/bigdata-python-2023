##### 정렬
- 정렬 기준: axis
    - 0: 인덱스
    - 1: 칼럼
- 정렬 방향: ascending
    - True: 오름차순
    - False: 내림차순

```python
# 데이터 포털에서 가져올 때 인덱스 기준 오름차순 정렬돼있음
df2
```

```python
# 인덱스 정렬 sort_index()
df2.sort_index(axis=0, ascending=False) # 인덱스 기준으로 내림차순 정렬
```

```python
df2.sort_index(axis=1, ascending=True) # 컬럼이름으로 오름차순 정렬
```

```python
# 값 정렬 sort_values(by='기준컬럼명')
df2.sort_values(by='노선')
```

##### 결측치 확인, 제거, 채우기
- 데이터분석 분석 이전에 하는 제일 중요한 일

```python
datas = {'이름' : ['길동이', '길순이', '유고', '애슐리'],
        '국어' : [90, 100, None, 99],
        '영어' : [None, 100, 99, 100],
        '수학' : [50, 100, None, 20]}

df5 = pd.DataFrame(datas)
df5
```

```python
df5.isnull().sum()
```

```python
df5.notnull().sum()
```

```python
# 결측치 채우기
# fillna
# inplace=True를 하면 별도로 저장하지 않아도 바로 확인가능함
df5['국어'] = df5['국어'].fillna(50)
df5
```

```python
df5['영어'].fillna(40, inplace=True)
df5
```

```python
# method='ffill' : forward 값으로 채워넣음
df5['수학'].fillna(method='ffill')
```

```python
# method='bfill' : backward 값으로 채워넣음
df5['수학'].fillna(method='bfill', inplace=True)
```

```python
df5
```

```python
# 결측치 삭제
df6 = pd.DataFrame(datas)
df6
```

```python
df6.dropna()
```

```python
df6
```

```python
df6.dropna(inplace=True)
df6
```

```python
df7 = pd.DataFrame(datas)
df7
```

```python
# column을 기준으로 None, NaN이 있는 column은 모두 삭제
df7.dropna(axis=1)
```

```python
# 형식 변환
df7['국어']
```

```python
df7['국어'].astype('object')
```

```python
# NaN이 있으면 float64 -> int64로 변경하려면 결측치가 없어야 변경 가능
df7['국어'].fillna(30).astype('int64')
```
