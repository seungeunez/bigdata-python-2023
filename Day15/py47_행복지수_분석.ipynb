```python
import matplotlib
# from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc, rcParams
import seaborn as sns

import pandas as pd
import numpy as np

font_path = 'C:/Windows/Fonts/NanumGothic.ttf'
font = font_manager.FontProperties(fname=font_path).get_name() # 'NanumGothic'
rc('font', family=font)
rcParams['axes.unicode_minus'] = False
```

```python
# 삶의 만족도 로드
df_happylife = pd.read_excel('./raw/대한민국행복지도_삶의만족도.xlsx')
df_happylife
```

```python
# 건강 데이터 로드
df_health = pd.read_excel('./raw/대한민국행복지도_건강.xlsx')
df_health
```

```python
# 안전
df_safety = pd.read_excel('./raw/대한민국행복지도_안전.xlsx')
df_safety
```

```python
# 환경
df_environ = pd.read_excel('./raw/대한민국행복지도_환경.xlsx')
df_environ
```

```python
# 경제
df_econ = pd.read_excel('./raw/대한민국행복지도_경제.xlsx')
df_econ
```

```python
# 교육
df_edu = pd.read_excel('./raw/대한민국행복지도_교육.xlsx')
df_edu
```

```python
# 관계 및 사회참여
df_relation = pd.read_excel('./raw/대한민국행복지도_관계 및 사회참여.xlsx')
df_relation
```

```python
# 여가
df_leisure = pd.read_excel('./raw/대한민국행복지도_여가.xlsx')
df_leisure
```

```python
# 시도별 데이터 추출
df_city = pd.DataFrame(df_happylife['시도'].unique())
df_city.columns = ['시도']
df_city
```

```python
# 삶의 만족도 병합
df_life = df_happylife['삶의 만족도'].groupby(by=df_happylife['시도']).mean() # 각 시도별 평균내서 17개로 그룹핑
df_happy_merge = pd.merge(df_city, df_life, how='left', on='시도')
df_happy_merge
```

```python
df_htemp = df_health['평균'].groupby(by=df_happylife['시도']).mean() # 각 시도별 평균내서 17개로 그룹핑
df_happy_merge = pd.merge(df_happy_merge, df_htemp.rename('건강'), how='left', on='시도')
df_happy_merge
```

```python
# 안전
df_temp = df_safety['평균'].groupby(by=df_happylife['시도']).mean() # 각 시도별 평균내서 17개로 그룹핑
df_happy_merge = pd.merge(df_happy_merge, df_temp.rename('안전'), how='left', on='시도')
df_happy_merge
```

```python
# 환경
df_temp = df_environ['평균'].groupby(by=df_happylife['시도']).mean() # 각 시도별 평균내서 17개로 그룹핑
df_happy_merge = pd.merge(df_happy_merge, df_temp.rename('환경'), how='left', on='시도')
df_happy_merge
```

```python
# 경제
df_temp = df_econ['평균'].groupby(by=df_happylife['시도']).mean() # 각 시도별 평균내서 17개로 그룹핑
df_happy_merge = pd.merge(df_happy_merge, df_temp.rename('경제'), how='left', on='시도')
df_happy_merge
```

```python
# 교육
df_temp = df_edu['평균'].groupby(by=df_happylife['시도']).mean() # 각 시도별 평균내서 17개로 그룹핑
df_happy_merge = pd.merge(df_happy_merge, df_temp.rename('교육'), how='left', on='시도')
df_happy_merge
```

```python
# 관계 및 사회참여
df_temp = df_relation['평균'].groupby(by=df_happylife['시도']).mean() # 각 시도별 평균내서 17개로 그룹핑
df_happy_merge = pd.merge(df_happy_merge, df_temp.rename('관계및사회참여'), how='left', on='시도')
df_happy_merge
```

```python
# 여가
df_temp = df_leisure['평균'].groupby(by=df_happylife['시도']).mean() # 각 시도별 평균내서 17개로 그룹핑
df_happy_merge = pd.merge(df_happy_merge, df_temp.rename('여가'), how='left', on='시도')
df_happy_merge
```

```python
df_happy_merge.to_excel('./대한민국_시도별_행복지수.xlsx', index=False)
```

##### 시각화

```python
# 선그래프 시각화
plt.figure(figsize=(15,6))
plt.title('대한민국 행복지수')
items = list(df_happy_merge.columns[1:9])
for item in items:
    data = df_happy_merge[item]
    plt.plot(data, marker='o', label=item)

plt.xlabel('시도')
plt.ylabel('수치')
plt.xticks(range(17), df_happy_merge['시도'], fontsize=8)
plt.legend() 
plt.grid()
plt.show()
```

```python
# 막대 그래프
df_happy_merge.plot(kind='bar', xlabel='시도', ylabel='수치', figsize=(15, 20), grid=True, subplots=True)

plt.suptitle('대한민국 행복지수', fontsize=20)
plt.tight_layout(pad=8, h_pad=2)
plt.xticks(range(17), df_happy_merge['시도'])
plt.show()
```

```python
# 히트맵
plt.figure(figsize=(15,13))
plt.title('대한민국 행복지수 상관관계', fontsize=20)

corr_mat = df_happy_value.corr()
upp_mat = np.triu(corr_mat)
sns.heatmap(corr_mat, mask=upp_mat, annot=True, cmap='RdYlGn')
plt.show()
```
